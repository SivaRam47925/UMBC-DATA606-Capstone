```python

# EDA NOTEBOOK (CODE-ONLY)
# Project: ICU Mortality Prediction (Proposal Stage)
# Author: Siva Ram Potluri
# Dataset: WiDS Datathon ICU Mortality (sample for GitHub size limit)


# 1) Imports & display settings

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 140)


# 2) Load dataset

# NOTE: This notebook uses a reduced sample (e.g., 20,000 rows) to keep file size under GitHub limits.
DATA_PATH = "/content/training_v2.csv"

df = pd.read_csv(DATA_PATH)

print(" Dataset loaded")
print("Path:", DATA_PATH)


# 3) Basic overview

print("\n--- BASIC OVERVIEW ---")
print("Shape (rows, cols):", df.shape)

print("\nFirst 5 rows:")
display(df.head())

print("\nLast 5 rows:")
display(df.tail())


# 4) Column info / data types

print("\n--- DATA TYPES & NON-NULL COUNTS ---")
df.info()

# Quick summary of data types
dtype_counts = df.dtypes.value_counts()
print("\nDtype counts:")
print(dtype_counts)

# Identify categorical-like columns (object)
cat_cols = df.select_dtypes(include=["object"]).columns.tolist()
num_cols = df.select_dtypes(exclude=["object"]).columns.tolist()

print(f"\nCategorical (object) columns: {len(cat_cols)}")
print(f"Numeric columns: {len(num_cols)}")

# 5) Target variable analysis

# IMPORTANT: Target label for ML = hospital_death (0 = survived, 1 = died)
TARGET_COL = "hospital_death"

print("\n--- TARGET VARIABLE DISTRIBUTION ---")
if TARGET_COL not in df.columns:
    raise ValueError(f"Target column '{TARGET_COL}' not found in dataset.")

target_counts = df[TARGET_COL].value_counts(dropna=False)
target_props = df[TARGET_COL].value_counts(normalize=True, dropna=False)

print("\nCounts:")
print(target_counts)

print("\nProportions:")
print(target_props)

# Plot target distribution
plt.figure(figsize=(6, 4))
sns.countplot(x=TARGET_COL, data=df)
plt.title("Target Distribution: hospital_death")
plt.xlabel("hospital_death (0 = Survived, 1 = Died)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# Brief imbalance note (proposal stage)
mortality_rate = target_props.get(1, np.nan) * 100
print(f"\nEstimated mortality rate in this sample: {mortality_rate:.2f}%")
print("Note: If the classes are imbalanced, later modeling should use metrics like ROC-AUC, F1, and Recall.")


# 6) Missing values analysis

print("\n--- MISSING VALUES ---")
missing_counts = df.isnull().sum()
missing_pct = (missing_counts / len(df)) * 100

missing_summary = pd.DataFrame({
    "missing_count": missing_counts,
    "missing_pct": missing_pct
}).sort_values("missing_count", ascending=False)

print("\nTop 25 columns with most missing values:")
display(missing_summary.head(25))

# Optional: how many columns have any missingness?
cols_with_missing = (missing_counts > 0).sum()
print(f"\nColumns with missing values: {cols_with_missing} / {df.shape[1]}")


# 7) Summary statistics (numeric)

print("\n--- NUMERIC SUMMARY STATISTICS ---")
display(df.describe(include=[np.number]).T)


# 8) Quick look at key features

# Pick a few intuitive variables (only if they exist in dataset)
key_features = ["age", "bmi", "pre_icu_los_days", "d1_heartrate_max", "d1_sysbp_min", "d1_lactate_max"]

existing_key_features = [c for c in key_features if c in df.columns]
print("\nExisting key features found:", existing_key_features)

# Histograms for key numeric variables
for col in existing_key_features:
    plt.figure(figsize=(6, 4))
    df[col].hist(bins=30)
    plt.title(f"Distribution: {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()

# 9) Categorical overview
print("\n--- CATEGORICAL OVERVIEW (TOP LEVELS) ---")
for col in cat_cols[:10]:  # show only first 10 categorical cols to keep output readable
    print(f"\nColumn: {col}")
    display(df[col].value_counts(dropna=False).head(10))


# 10) Preliminary notes for modeling (proposal stage)

print("\n--- PRELIMINARY NOTES ---")
print("1) Target = hospital_death")
print("2) Handle missing values (median for numeric, most-frequent for categorical)")
print("3) Encode categorical variables (one-hot or target encoding later)")
print("4) Evaluate with ROC-AUC + Recall/F1 due to imbalance")
print("5) Consider comparing models with vs without APACHE-related features (to reduce leakage risk)")
```
